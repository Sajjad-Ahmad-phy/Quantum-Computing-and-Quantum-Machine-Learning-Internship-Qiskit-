```python
from qiskit.visualization import plot_bloch_multivector as plt
import numpy as np

a_0 = np.array([1, 0])
a_1 = np.array([0,1])

plt(a_0, title='1 Qubit state', figsize = (6, 4))
```

```python
from qiskit.visualization import plot_bloch_multivector as plt
import numpy as np

a_0 = np.array([1, 0])
a_1 = np.array([0,1])

qs = a_0 + 3j*a_1      #quantum state

mag = np.sqrt(abs(qs[0]**2) + abs(qs[1]**2))
state = qs/mag

plt(state, title='1 Qubit state', figsize = (6, 4))
```
